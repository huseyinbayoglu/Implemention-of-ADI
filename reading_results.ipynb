```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("Results.csv")

fig, ax = plt.subplots(2, 2, figsize=(10, 8))

# Policy Loss
ax[0, 0].plot(df["policy_output_loss"][1:], "r", label="Policy Loss")
ax[0, 0].set_title("Policy Loss")
ax[0, 0].legend()
ax[0, 0].grid(True)
ax[0, 0].set_ylim([min(df["policy_output_loss"][1:]) * 0.9, max(df["policy_output_loss"][1:]) * 1.1])

# Policy Accuracy
ax[0, 1].plot(df["policy_output_accuracy"][1:], "r", label="Policy Accuracy")
ax[0, 1].set_title("Policy Accuracy")
ax[0, 1].legend()
ax[0, 1].grid(True)
ax[0, 1].set_ylim([0, 1.1])  # Accuracy genelde 0-1 arasnda olur

# Value Loss
ax[1, 0].plot(df["value_output_loss"][1:], "b", label="Value Loss")
ax[1, 0].set_title("Value Loss")
ax[1, 0].legend()
ax[1, 0].grid(True)
ax[1, 0].set_ylim([min(df["value_output_loss"][1:]) * 0.9, max(df["value_output_loss"][1:]) * 1.1])

# Value MSE
ax[1, 1].plot(df["value_output_mse"][1:], "b", label="Value MSE")
ax[1, 1].set_title("Value MSE")
ax[1, 1].legend()
ax[1, 1].grid(True)
ax[1, 1].set_ylim([min(df["value_output_mse"][1:]) * 0.9, max(df["value_output_mse"][1:]) * 1.1])

# Genel ayarlar
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of steps
n_steps = 500_000

# Generate random steps: each element is either +1 or -1
steps = np.random.uniform(-1, 1, size=n_steps)

# Compute the cumulative sum to get the position at each step
positions = np.cumsum(steps)

# Insert the starting position (0) at the beginning
positions = np.insert(positions, 0, 0)

# Plot the random walk
plt.figure(figsize=(10, 6))
plt.plot(positions, label='Random Walk')
plt.title(f'Random Walk (+1/-1) over {n_steps} steps')
plt.xlabel('Step')
plt.ylabel('Position')
plt.grid(True)
plt.legend()
plt.show()
np.max(positions), np.min(positions)
```
